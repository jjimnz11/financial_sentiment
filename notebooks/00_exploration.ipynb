# 00 — Exploración de titulares financieros

**Objetivo:** Entender la distribución de noticias scrapeadas antes de pasar por el pipeline NLP.

- ¿Cuántos titulares por fuente y por día?
- ¿Hay gaps temporales?
- ¿Cómo se distribuyen las palabras clave?
- ¿Correlación visual entre volumen de noticias y VIX?

```python
import sys
sys.path.insert(0, '..')  # Permite imports desde src/

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

from src.utils.storage import HeadlineCache
from src.utils.config_loader import load_config

config = load_config()
cache = HeadlineCache()
print(f"Total headlines en cache: {cache.count():,}")
```

```python
# Carga los últimos 30 días
since = datetime.utcnow() - timedelta(days=30)
df = cache.get_since(since)
print(df.shape)
df.head()
```

```python
# Distribución por fuente
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

df['source'].value_counts().plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Titulares por fuente')
axes[0].set_xlabel('')

# Volumen diario
df['date'] = pd.to_datetime(df['published']).dt.date
df.groupby('date').size().plot(ax=axes[1], color='coral')
axes[1].set_title('Titulares por día')

plt.tight_layout()
plt.savefig('../outputs/figures/00_headline_distribution.png', dpi=150)
plt.show()
```
